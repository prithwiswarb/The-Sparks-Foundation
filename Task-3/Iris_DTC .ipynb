### 1. Importing required packages.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#ignore the warning messages.
import warnings
warnings.simplefilter("ignore")
```

### 2. Importing  the dataset and extracting the facts.

```python
iris=pd.read_csv("Iris.csv") #creating dataframe of the csv file
```

```python
iris.columns
```

```python
iris.info()
```

```python
iris['Species'].value_counts()
```

```python
iris=iris.drop('Id',axis=1)# Id package is not required.
```

### 3. Pre-processing the Dataset.

```python
X = iris.iloc[:, :-1].values
y = iris.iloc[:, 4].values
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import graphviz
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

### 4. Applying Decision Tree Classifier.

```python
clf = tree.DecisionTreeClassifier(random_state=17)
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

### 5. Accuracy of Algorithm.

```python
print('\nAccuracy: {0:.4f}'.format(accuracy_score(y_test, y_pred)))
```

### 6. Visualising Decision Tree Classifier Graphically.

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data=export_graphviz(clf,out_file=None,filled=True,proportion=True,
                        feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],
                        class_names=['Setosa','Versicolor','Virginica'])
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
